####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# 1) Função em Python

def filtro_uf_tipo(uf: str, tipo_interessado: str):
    url_completa = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    
    response = requests.get(url_completa)
    data = response.json()
    df = pd.DataFrame(data['items'])

    # Filtrando UF e tipo interessado (Estado ou Município)
    filtro = (df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)
    df_filtrado = df[filtro]

    return df_filtrado
````

````python
# 2) Consulta de status em 'Arquivado por decurso de prazo' para MG

uf = 'MG'
tipo_interessado = 'Estado'

df_filtrado = filtro_uf_tipo(uf, tipo_interessado)

df_arquivado = df_filtrado[df_filtrado['status'] == 'Arquivado por decurso de prazo']

quantidade_arquivado = df_arquivado.shape[0]

print(f"A quantidade de status 'Arquivado por decurso de prazo' para {tipo_interessado} de {uf} é: {quantidade_arquivado}")
````

````python
# 3) Município da Bahia com mais solicitações deferidas

uf = 'BA'
tipo_interessado = 'Município'
url_completa = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'

# Requisição à API
response = requests.get(url_completa)
data = response.json()
df = pd.DataFrame(data['items'])

# Verificar as colunas disponíveis no DataFrame
# print(df.columns)

#Filtro para deferidos
df_deferido = df[df['status'] == 'Deferido']

municipio_mais_frequente = df_deferido['interessado'].value_counts().head(1)

print(f"O município da Bahia com mais solicitações deferidas é:")
print(municipio_mais_frequente)
````

````python
# 4) Salvando os dados de solicitações da Bahia, com interessado = 'Estado' em um arquivo .csv

uf = 'BA'
tipo_interessado = 'Estado'
url_completa = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'

# Requisição à API e criação do DataFrame
response = requests.get(url_completa)
data = response.json()
df_bahia_estado = pd.DataFrame(data['items'])

# Salvando o DataFrame em um arquivo CSV
nome_arquivo = 'solicitacoes_bahia.csv'
df_bahia_estado.to_csv(nome_arquivo, index=False) # index=False evita que o índice numérico do DataFrame seja salvo como uma coluna no CSV.

# Conferindo as primeiras linhas do DataFrame que foi salvo
print("\nPrimeiras 5 linhas do DataFrame salvo:")
print(df_bahia_estado.head())
````
